# Matplotlib
Matplotlib is a Python 2D ploting library which produces publication-quality fgures in a variety of hardcopy formats and interactive environments across platforms.
___

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Preparing Data

import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.grid = 1
plt.plot(x,z)
plt.xlabel("range")
plt.ylabel("sin")
plt.axis([0, 6, -1, 1])
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
